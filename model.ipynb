```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split 
```

```python
data=pd.read_csv("D:\\DATASETS\\Admission_Predict.csv")
data.head()
```

```python
max(data['LOR '])
```

```python
data.columns
```

```python
data.shape
```

```python
data=data.drop('Serial No.',axis=1)
```

```python
data.isnull().sum()
```

```python
X=data.drop('Chance of Admit ',axis=1)
y=data['Chance of Admit ']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=56)
```

```python
X_train
```

```python
num_cols=['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
        'Research']
```

```python
num_cols
```

```python
num_pip=Pipeline(steps=[
    ('scaler',StandardScaler())
])
```

```python
preprocessor=ColumnTransformer(transformers=[
    ("num",num_pip,num_cols)
])
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
model = Pipeline([
    ('pre', preprocessor),  
    ('sgd', SGDRegressor(learning_rate='constant', eta0=0.01, max_iter=1000, tol=1e-3)) 
])
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print("r2 Score:",r2_score(y_test,y_pred))
print("Mean Absolute Error:",mean_absolute_error(y_test,y_pred))
print("Mean Absolute Error:",mean_squared_error(y_test,y_pred))
```
